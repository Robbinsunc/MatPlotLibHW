## Pyber Observable Trends:

#### Urban Cities provide 80.1% of the total number of drivers, provide 68.4% of the total number of rides and accounts for 62.7% of the total fares.
#### Suburan Cities provide 16.5% of the total number of drivers, provide 26.3% of the total number of rides and accounts for 30.5% of the total fares.
#### Rural Cities provide 2.6% of the total number of drivers, provide 5.3% of the total number of rides and accounts for 6.8% of the total fares.

#### Based on the Pyber Riding Sharing Data (2016) bubble plot, as a general trend, there is an inverse correlation between the total number of rides per city and the average fare per city. As the city type moved from Rural to Urban the number of rides per city tended to increase and the average fare per city tended to decrease. This makes sense given one would expect a more densely populated area such as an Urban city to have more customers with shorter ride times and a higher level of competition to help keep prices down. The opposite would be true of a more Rural city setting with less customers and longer ride times leading to a higher fare per city. This analysis do not factor for cost of living. Secondly, there is also inverse correlation between average number of drivers per city and the average fare per city. As the number of drivers per city increases, the average fare per city tends to decrease. Which again would follow along with the first trend of larger Urban cities having a greater population and shorter drive times leading to lower average fares while Rural less populated areas would have lower populations and longer drive times leading to a higher average fare per city. Lastly, although Rural areas tend to have a higher average fare per city, Rural areas account for 6.8% of the total fares and while Urban cities tend to have a lower average fare per city, Urban cities account for 62.7% of the total fares. Suburan areas account for the difference of 30.5% of the total fares. This data shows the ride sharing business relies on having a high number of rides. Although the average fare per city might be lower, the total revenue from rides far exceedes having less rides and a higer average fare per city. 

```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data = "data/city_data.csv"
ride_data = "data/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data)
ride_data_df = pd.read_csv(ride_data)

#city_data_df.head()
#ride_data_df.head()
# Combine the data into a single dataset
pyber_df = pd.merge(ride_data_df, city_data_df, on='city', how='outer')

# Display the data table for preview
pyber_df.head()
```

```python
#groupby city
pyber_groupby = pyber_df.groupby('city')
pyber_groupby.count().head()
```

```python
#find average fare from groupby df
ave_fare = pyber_groupby['fare'].mean()
ave_fare.head()
```

```python
#find number of rides per city from groupby df
num_rides = pyber_groupby['city'].count()
num_rides.head()
```

```python
#find number of drivers using group by df
num_drivers = pyber_groupby['driver_count'].mean()
num_drivers.head()
```

```python
#pass through city type from city_datadf to new df
city_type_df = city_data_df.set_index('city')
city_type = city_type_df['type']
city_type.head()
```

```python
#Create new DataFrame for new variables
pyber_data = pd.DataFrame({'Average Fare ($) Per City': ave_fare, 'Total Number of Rides Per City': num_rides,
                           'Total Number of Drivers Per City': num_drivers, 'City Type': city_type})
pyber_data.head()
```

```python
# Obtain the x and y coordinates for each of the three city types
#rural
rural = pyber_data.loc[pyber_data['City Type']=='Rural']
#rural.head()
#suburban
suburban = pyber_data.loc[pyber_data['City Type']=='Suburban']
#suburban.head()
#urban
urban = pyber_data.loc[pyber_data['City Type']=='Urban']
#urban.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Build the scatter plots for each city types
# x = Average Fare; y= Total Number of Rides
fig, ax=plt.subplots()
rural.plot(kind='scatter', x='Total Number of Rides Per City', y='Average Fare ($) Per City', ax=ax, color='gold', 
           edgecolors='black', s = rural['Total Number of Drivers Per City'] *10, alpha = 0.75, linewidth = 1, 
          label = 'Rural')
suburban.plot(kind='scatter', x='Total Number of Rides Per City', y='Average Fare ($) Per City', ax=ax, color = 'lightskyblue', 
              edgecolors='black', s = suburban['Total Number of Drivers Per City'] *10, alpha = 0.75, linewidth = 1, 
             label = 'Suburban')
urban.plot(kind='scatter', x='Total Number of Rides Per City', y='Average Fare ($) Per City', ax=ax, color='lightcoral', 
           edgecolors='black', s = urban['Total Number of Drivers Per City'] *10, alpha = 0.75, linewidth = 1, 
          label = 'Urban')
# Incorporate the other graph properties
ax.grid(alpha = 0.75)
plt.xlim(left=0, right=41)
plt.ylim(bottom=19)
plt.title('Pyber Ride Sharing Data (2016)')
plt.ylabel('Average Fare ($)')

# Create a legend
lgnd = plt.legend(loc='best', title = 'City Types')
for handle in lgnd.legendHandles:
    handle.set_sizes([44.0])

# Incorporate a text label regarding circle size
plt.text(42, 36.5, 'Note:')
plt.text(42, 35, 'Circle size correlates with driver count per city')

# Save Figure
plt.savefig('../Pyber/Pyber_Ride_Sharing_Plot.png', bbox_inches='tight')
#show plot
```

## Total Fares by City Type

```python
# Calculate Type Percents
city_type_fare = pyber_df.groupby(['type'])
city_type_total_fare = city_type_fare['fare'].sum()
city_type_total_fare
```

```python
# Build Pie Chart
colors = ['gold', 'lightskyblue', 'lightcoral']
explode = (0, 0, 0.10)
city_type_total_fare.plot(kind='pie', colors=colors, explode=explode, autopct ='%1.1f%%', startangle=150, shadow=True)
plt.title("% of Total Fares by City Type")
plt.ylabel('')
#show Figure
# Save Figure
plt.savefig('../Pyber/Pyber_FaresbyCityType.png')
```

## Total Rides by City Type

```python
# Calculate Ride Percents
city_type_rides = pyber_df.groupby(['type'])
city_type_total_rides = city_type_rides['ride_id'].count()
city_type_total_rides
```

```python
# Build Pie Chart
colors = ['gold', 'lightskyblue', 'lightcoral']
explode = (0, 0, 0.10)
city_type_total_rides.plot(kind='pie', colors=colors, explode=explode, autopct ='%1.1f%%', startangle=150, shadow=True)
plt.title("% of Total Rides by City Type")
plt.ylabel('')
#show Figure
# Save Figure
plt.savefig('../Pyber/Pyber_RidesbyCityType.png')
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
city_type_drivers = pyber_data.groupby(['City Type'])
city_type_total_drivers = city_type_drivers['Total Number of Drivers Per City'].sum()
city_type_total_drivers
```

```python
# Build Pie Charts
colors = ['gold', 'lightskyblue', 'lightcoral']
explode = (0, 0, 0.10)
city_type_total_drivers.plot(kind='pie', colors=colors, explode=explode, autopct ='%1.1f%%', startangle=165, shadow=True)
plt.title("% of Total Drivers by City Type")
plt.ylabel('')
#show Figure
# Save Figure
plt.savefig('../Pyber/Pyber_DriversbyCityType.png')
```
